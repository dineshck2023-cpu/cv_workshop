```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Download image using wget (works in Colab)
# This line uses a shell command (!) to download an image from the internet
# The image will be saved as "sample.jpg" in your current working directory

!wget -O sample.jpg https://images.unsplash.com/photo-1518791841217-8f162f1e1131

# Read and display
# Import OpenCV library (cv2) for image processing
import cv2

# Import matplotlib for displaying the image
import matplotlib.pyplot as plt
# Read the image from the file "sample.jpg"
# cv2.imread() loads the image in BGR (Blue, Green, Red) format by default
image = cv2.imread("sample.jpg")
# Convert the image from BGR format to RGB format
# This is important because matplotlib expects images in RGB format
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Display the image using matplotlib
plt.imshow(image)
# Add a title to the displayed image
plt.title("Original Image")
# Turn off axis (removes x and y axis numbers for a cleaner look)
plt.axis("off")
```

```python
print("Pixel value at (100,100): ", image[100][100])
```

convert to grayscale

```python
# Convert the original RGB image into a grayscale image
# Grayscale means the image will have only intensity (brightness) values
# No colors – just shades of black, white, and gray
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Display the grayscale image using matplotlib
# cmap='gray' tells matplotlib to show the image in grayscale format
plt.imshow(gray, cmap='gray')

# Add a title to the image
plt.title("Grayscale Image")

# Turn off axis (removes x and y axis numbers for a cleaner view)
plt.axis('off')
```

```python
# Import required libraries
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('sample.jpg') # Changed from 'image.jpg' to 'sample.jpg'

# Convert BGR to RGB for correct display
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Get image dimensions
# image.shape returns (height, width, channels)
h, w, _ = image.shape

# Crop center region safely (no out-of-bounds)
# Select rows from 20% to 70% of height
# Select columns from 20% to 70% of width
crop = image[int(h*0.2):int(h*0.7), int(w*0.2):int(w*0.7)]

# Display the cropped image
plt.imshow(crop)

# Add a title
plt.title("Cropped Image")

# Turn off axis for cleaner look
plt.axis('off')

# Show output
plt.show()
```

```python
# Import required libraries
import cv2
import matplotlib.pyplot as plt

# Load the image (replace with your image path)
image = cv2.imread('sample.jpg') # Changed from 'image.jpg' to 'sample.jpg'

# Convert BGR (OpenCV default) to RGB (for correct display)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Resize the image to 300x300 pixels
resized = cv2.resize(image_rgb, (300, 300))

# Display the resized image
plt.imshow(resized)

# Add title
plt.title("Resized Image")

# Turn off axis
plt.axis('off')

# Show the image
plt.show()
```

edge detection

```python
# Apply Canny Edge Detection
# 100 = lower threshold, 200 = upper threshold
edges = cv2.Canny(gray, 100, 200)

# Display the edges
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')

# Remove axis for clean output
plt.axis('off')
```
